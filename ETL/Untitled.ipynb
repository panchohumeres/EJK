```python
date = '2020-05-05'
```

```python

dataTotalRegion = pd.read_csv("https://raw.githubusercontent.com/MinCiencia/Datos-COVID19/master/output/producto4/"+date+"-CasosConfirmados-totalRegional.csv", index_col='Region')
```

```python
df=dataTotalRegion 
```

```python
index=df.columns.to_list()
for cidx in range(len(df.columns)):
    if '%' in df.columns[cidx]:
        index[cidx]=index[cidx].replace('%','Porcentaje')
        df.columns=index
        col=index[cidx]
        df[col]=df[col].str.replace('%','').astype(float)
    if "*" in df.columns[cidx]:
        index[cidx]=index[cidx].replace('*','')
        df.columns=index
```
